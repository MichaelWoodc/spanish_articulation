foreach ($f in Get-ChildItem -Recurse -Include *.mov, *.mp4) {
    $out = "$($f.DirectoryName)\$($f.BaseName)_360p.mp4"
    ffmpeg -i "$($f.FullName)" -vf "scale=-2:360" -c:v libx264 -preset veryslow -crf 28 -c:a aac -b:a 64k -movflags +faststart "$out"
}


```python
import subprocess

ps_script = r'''
Get-ChildItem -Recurse -Include *.mov, *.mp4 | ForEach-Object {
    $out = "$($_.DirectoryName)\$($_.BaseName)_360p.mp4"
    ffmpeg -i "`"$($_.FullName)`"" -vf "scale=-2:360" -c:v libx264 -preset veryslow -crf 28 -c:a aac -b:a 64k -movflags +faststart "`"$out`""
}
'''

process = subprocess.Popen(
    ["powershell", "-NoProfile", "-Command", ps_script],
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT,
    text=True,
    bufsize=1
)

# Stream output in real time
for line in process.stdout:
    print(line, end="")

process.wait()
```
